# Probability distributions of interactions during MD simulations

## Base-specific hydrogen bonds


```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import csv
import os

input_files = os.path.join('DNMT3B_inputs')

a=pd.read_csv(input_files+"/3B-CGC-countbasespechbond", sep="\t", header=None )
a.columns = ["count"]
a['Type'] = 'CGC'

b=pd.read_csv(input_files+"/3B-CGG-countbasespechbond", sep="\t", header=None )
b.columns = ["count"]
b['Type'] = 'CGG'

c=pd.read_csv(input_files+"/3B-CGT-countbasespechbond", sep="\t", header=None )
c.columns = ["count"]
c['Type'] = 'CGT'

d=pd.read_csv(input_files+"/3B-CGA-countbasespechbond", sep="\t", header=None )
d.columns = ["count"]
d['Type'] = 'CGA'

dnmt3b_bshb = pd.concat([b, a, c, d], ignore_index=True)
dnmt3b_bshb
```

```python
import numpy as np

plt.figure(figsize=(7, 6))

palet=['#746AB0','#A9A9A9', 'orange', 'lightseagreen']

sns.violinplot(x="Type", y="count", data=dnmt3b_bshb,bw=0.5,cut=0, scale='count', palette=palet)

plt.xticks(fontsize=13, fontweight='bold')
plt.yticks(fontsize=13, fontweight='bold')
plt.yticks(np.arange(0, 19, 3))
plt.ylabel('Number of h-bonds',fontsize=15, fontweight='bold')
plt.ylim(ymax = 20, ymin = -2)

plt.xlabel('',fontsize=13, fontweight='bold')
plt.title("Base-specific h-bond - DNMT3B",fontsize=20, fontweight='bold')

plt.savefig("3b-cpg-bshb-violin-label.pdf", dpi=600, bbox_inches='tight', format="pdf")
```
